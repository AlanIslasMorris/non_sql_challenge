5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

update_query = [
    {"$set": {"geocode.latitude": {"$toDouble": "$geocode.latitude"}}},
    {"$set": {"geocode.longitude": {"$toDouble": "$geocode.longitude"}}}
]

result = establishments.update_many({}, update_query)
```

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : "3"}} ])
```

```python

from bson.objectid import ObjectId

# Retrieve the documents to be updated
documents = establishments.find({})

# Update each document individually


for document in documents:
    document_id = document["_id"]
    if "RatingValue" in document:
        rating_value = document["RatingValue"]
        updated_rating_value = int(rating_value)
        
        establishments.update_one(
            {"_id": ObjectId(document_id)},
            {"$set": {"RatingValue": updated_rating_value}}
        )
```

```python

# Perform aggregation to get the unique RatingValue values
criteria = [
    {"$group": {
        "_id": None,
        "unique_rating_values": {"$addToSet": "$RatingValue"}
    }}
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(criteria))

# Extract the unique RatingValue values
print(result)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Check that the coordinates and rating value are now numbers

criteria = [
    {"$group": {
        "_id": None,
        "mean_rating": {"$avg": "$RatingValue"},
        "mean_latitude": {"$avg": {"$toDouble": "$geocode.latitude"}},
        "mean_longitude": {"$avg": {"$toDouble": "$geocode.longitude"}}
    }}
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(criteria))

print(result)
```
